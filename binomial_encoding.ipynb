# Nhập môn SAT Solving với PySAT

## I. NP vs P

**1. Thời gian đa thức (Polynomial time)**
> Một thuật toán được gọi là hoàn thành trong thời gian đa thức nếu số bước thực thi của thuật toán cho tới tìm ra lời giải là O(n^k) với k > 0. Các phép toán cơ bản +, - , x, :, log được thực hiện trong thời gian đa thức 

**2. NP**
> là tập hợp các bài toán mà lời giải có thể kiểm tra được trong thời gian đa thức
   - VD: Tìm ra tập con khác rỗng có tổng bằng 0 trong một mảng số nguyên
     + Chưa có lời giải nào giải trong thời gian đa thức
     + Duyệt tất cả các tập con. Số cách chọn 1 tập con là: 2^n (vì mỗi phần tử có 2 trạng thái 1 là có ở tập con 2 là không có ở tập con = 2 * 2 * ... * 2).

**3. P**
>Là tập hợp các bài toán có lời giải có thể thực hiện trong thời gian đa thức
- VD: Giải phương trình bậc 2, tìm ước chung lớn nhất, bội chung nhỏ nhất, ...

**4. NP-Hard**
> Là tập hợp các bài toán có thời gian giải ngoài thời gian đa thức và thời gian kiểm tra cũng chưa chắc nằm trong thời gian đa thức
- VD:
   + cực khó giải, cực khó kiểm tra: Tìm nước đi tối ưu trong một bàn cờ vua kích thước n * n
   + cực khó giải, vẫn kiểm tra được: bài toán soduku ở kích thước lớn
     * Làm không biết bao giờ xong vì giả sử là kích thước là 1000 x 1000
     * Nhưng kiểm tra thì nhanh với một bản hoàn thành ai đó đưa cho chỉ duyệt xem có số nào sai quy luật hay không

5. NP-Complete
> Là giao của NP với HP-Hard gồm các bài toán có thời gian giải ngoài thời gian đa thức và than gian kiểm tra nằm trong thời gian đa thức. Là những bài toán khó nhất trong các bài toán thuộc lớp NP
- VD:
  + Sudoku n * n
  + Tìm chu trình hamilton

## II. Bài toán SAT

> Cho công thức logic mệnh đề CNF, xác định xem công thức đó có thủa mãn hay không

VD: (x1 ∨ ¬x2) ∧ (x2 ∨ x3) có thỏa mãn được vì:

- x1 = true, x2 = false, x3 = true
- x1 = true, x2 = true, x3 = true

> Độ phức tạp: SAT là bài toán NP-complete \
> (NP = Nondeterministic Polynomial time)

Đây là bài toán NP-complete vì:
- Tìm ra lời giải cực kì khó nếu có 1000 biến thì có 2^1000 cách thử khác nhau
- Nhưng nếu có một lời giải cho trước thì kiểm tra cực kì nhanh

> Ưng dụng: Kiểm chứng phần mềm (Soft verification), AI, Planning 

## III. SAT Solver và PySAT

**2. PySAT**

## IV. Sử dụng PySAT với Glucose3 (skibidi)

```python
from pysat.solvers import Glucose3
# create a SAT Slover instance (glucose3)
with Glucose3() as g:
    #Add clauses to the solver
    g.add_clause([1, -2]) #x1 ∨ ¬x2
    g.add_clause([2, 3]) #x2 ∨ x3

    if g.solve():
        print("SATISFIABLES")
        print("Model: ", g.get_model()) #Get the model (Satisfying assignment) TTT
    else:
        print("UNSATISFIABLES")
```

## V. Bài toán N-Queen

**1. Code**
- Dùng set là một hashtable để kiểm tra có trùng cột, hành, đường chéo nhanh hơn ~O(1). Thay vì dùng mảng mất O(n) để kiểm tra

```python
def solve(n):
    pos = [-1] * n

    used_col = set()
    diag1 = set()  # row - col
    diag2 = set()  # row + col

    def backtrack(row):
        if row == n:
            print_board(pos)
            return
        
        for col in range(n):
            if col in used_col:
                continue
            
            if (row - col) in diag1:
                continue
            
            if (row + col) in diag2:
                continue

            pos[row] = col
            used_col.add(col)
            diag1.add(row - col)
            diag2.add(row + col)

            backtrack(row + 1)


            used_col.remove(col)
            diag1.remove(row - col)
            diag2.remove(row + col)
            pos[row] = -1

    backtrack(0)
```

```python
def print_board(pos):
    n = len(pos)
    for i in range(n):
        row = ['.'] * n
        row[pos[i]] = 'Q'
        print(' '.join(row))
```

```python
solve(4)
```

**2. Các dàng buộc phổ biến trong lập trình dàng buộc**

- Ít nhất một(`At least one`) = tổn tại: ALO(x1, x2, ...)
    + x1 ∨ x2 ∨ ...
- Nhiều nhất một (`At most one`): AMO(x1, x2, ...),
    + ¬(xi ∧ xj) = ¬xi ∧ ¬xj (i != j): đúng khi 1 trong hai cái không cùng True
- Chính xác một (`Exactly one`): EO(x1, x2, ...)
    + ALO(x1, x2, ...) ∨ AMO(x1, x2, ...)


**3. N-Queens: Biểu diễn ràng buộc với Pairwise Encoding**

a. Ràng buộc 1: Mỗi hàng có `Exactly One` quân hậu
- EO(xi1, xi2, ...)
- CNF
  + (xi1 ∨ xi2 ∨ ... ∨ xiN )
  + Với mỗi cặp k1, k2 khác nhau: (¬xik1 ∨ ¬xik2)

b. Ràng buộc 2: Mỗi cột có `Exactly One` quân hậu (tương tự RB1)

c. Ràng buộc 3: Mỗi đường chéo có `At Most One` quân hậu
- Với mỗi cặp ô (i, j) và (k, l) trên cùng một đường chéo (i != k, j != l,
|i-k| = |j-l|): (¬xij ∨ ¬xkl)

**4. N-Queen (N = 4) với PySAT và Pairwise Encoding I**

```python
def solve_nqueens(n):
    with Glucose3() as g:
        vars = [[(i * n + j + 1) for j in range(n)] for i in range (n)]
        #Exactly one for each row
        
        for i in range(n):  # At least one
            g.add_clause(vars[i])
            
            for j in range(n): #At most one
                for k in range(j + 1, n):
                    g.add_clause([-vars[i][j], -vars[i][k]])  

        #Exactly One for each column
        
        for j in range(n): # At least one
            g.add_clause([vars[i][j] for i in range(n)])
            
            for i in range(n): #At most one
                for k in range(i + 1, n):
                    g.add_clause([-vars[k][j], -vars[i][j]])
                    
        # At Most One for each diagonal
        for i1 in range(n):
            for j1 in range(n):
                for i2 in range (n):
                    for j2 in range(n):
                        if (i1 != i2 and j1 != j2 and abs ( i1 - i2 ) == abs( j1 - j2 )):
                            g.add_clause([-vars[i1][j1], -vars[i2][j2]])
        if g.solve():
           model = g . get_model () 
           for i in range (n) :
                row = ['.'] * n
                for j in range (n):
                    if vars[i][j] in model: row[j] = 'Q'
                print(' '.join(row))
        else :
           print (" UNSATISFIABLE ")
```

```python
solve_nqueens(10)
```

## BTVN

### Quy bài Sudoku về bài SAT

#### Tư duy: 
- Nếu như ở bài con N-Queens dùng mảng 2 chiều để mô tả mệnh đề $vars[i][j]$: con hậu ở vị trí (i, j) thì bải Sodoku phải tăng thêm 1 chiều vì 
    + Không chỉ đặt số mà còn phải chọn số để đặt
    + $vars[i][j][k]$: số k đặt ở vị trí i, j

#### Constrains 

- Mỗi ô chỉ chứa một số: $ EO(vars[i][j][0] ,..., vars[i][j][8], ... )$
- Mỗi hàng chỉ có một số k: $EO( vars[i][0][k] ,..., vars[i][8][k], ... )$
- Mỗi cột chỉ có một số k: $EO( vars[0][j][k] ,..., vars[8][j][k], ... )$
- Mỗi ô vuông 3 x 3 ... chỉ có một số k: $EO( vars[i + del//sqrt(n)][j + del \% sqrt(n)][k] ,..., vars[i + + del//sqrt(n)][j + del \%sqrt(n)][k], ... )$

#### Code

```python
import math
from pysat.solvers import Glucose3 
```

```python
def solvers_sudoku(n, grid):
    g = Glucose3()
    
    s = int(math.sqrt(n))
    vars = [[[n*n*i + n*j + k + 1 for k in range (n)] for j in range (n)] for i in range (n)]
    for i in range(n):
        for j in range(n):
            if grid[i][j] != 0:
                g.add_clause([vars[i][j][grid[i][j] - 1]])
    #exactly one for one cell
    for i in range (n):
        for j in range (n):
            g.add_clause(vars[i][j])
            for k1 in range (n - 1):
                for k2 in range (k1 + 1, n):
                    g.add_clause([-vars[i][j][k1], -vars[i][j][k2]])
    #exactly one for each row
    for k in range(n):
        for i in range(n):
            g.add_clause([vars[i][j][k] for j in range (n)])
            for j1 in range (n):
                for j2 in range (j1 + 1, n):
                    g.add_clause([-vars[i][j1][k], -vars[i][j2][k]])
    #exactly one for each col
    for k in range(n):
        for j in range(n):
            g.add_clause([vars[i][j][k] for i in range (n)])
            for i1 in range (n):
                for i2 in range (i1 + 1, n):
                    g.add_clause([-vars[i1][j][k], -vars[i2][j][k]])

    #exactly one for cell sqrt(n) * sqrt(n)
    for k in range (n):
        for i in range (0, n, s):
            for j in range (0, n, s):
                g.add_clause([vars[i + d//s][j + d%s][k] for d in range (n)]) #(d//s, d%s) là toạn độ của ma trận 3 * 3, i, j là dịch ma trận 
                for id1 in range(n):
                    for id2 in range (id1 + 1, n):
                        g.add_clause([ -vars[i + id1//s][j + id1%s][k],
                                        -vars[i + id2//s][j + id2%s][k]])

    if g.solve():
        model = g.get_model()
        for i in range(n):
            for j in range(n):
                for k in range(n):
                    if vars[i][j][k] in model:
                        print(k+1, end = ' ')
            print("\n")
    else:
        print("UNSATISFIABLE")
```

```python
grid = [
    [5,3,0,0,7,0,0,0,0],
    [6,0,0,1,9,5,0,0,0],
    [0,9,8,0,0,0,0,6,0],
    [8,0,0,0,6,0,0,0,3],
    [4,0,0,8,0,3,0,0,1],
    [7,0,0,0,2,0,0,0,6],
    [0,6,0,0,0,0,2,8,0],
    [0,0,0,4,1,9,0,0,5],
    [0,0,0,0,8,0,0,7,9],
]
```

```python
solvers_sudoku(9, grid)
```
