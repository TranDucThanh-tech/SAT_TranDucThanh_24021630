# Sequential counter encoding

## I. AMO (At most one)

**1. AMO** 
- $X_1 + X_2 + X_3 + ... + X_n \le 1 $ 

**2. Encoding**

- **Ý tưởng:**
  + Sử dụng biến $S_i$ để lưu trạng xem từ $X_i$ trở về trước có `ít nhất` $1$ biến $True$ hay chưa. 
  + Để tổng $X_1 + X_2 + ... + X_n \le 1 
    \leftrightarrow$ Không bao giờ để `prefix vượt quá 1`
    > $\forall i:\ \sum_{t=1}^{i} x_t \le 1$
- **Biến:**
  + n biến chính $X_i (0 < i \le n)$
  + n - 1 biến phụ $S_i  (0 < i < n)$
    * th1: $\sum_{j=1}^{i} X_j \ge 1 \rightarrow S_i = 1$
    * th2: $\sum_{j=1}^{i} X_j \le 0 \rightarrow S_i = 0$
    > $S_i$ là True nếu tổn tại $X_j$ là True mà $j \le i$
- **Ràng buộc:**
    + `Khởi tạo:`
        * $X_1 \rightarrow S_1$
          > Thiết lập trạng thái ban đầu: Nếu biến đầu tiên đúng thì hệ thống bắt đầu ở mức 1
    + `Duy trì trạng thái:`
        * $X_i \rightarrow S_i (1 < i < n)$
        * $S_{i-1} \rightarrow S_i (1 < i < n)$
          > Duy trì trạng thái "đã có ít nhất một" qua các bước từ $i=2$ đến $n-1$.
    + `Ràng buộc loại trừ:`
        * $S_{i-1} \rightarrow \neg X_i (1 < i \le n)$
          > Đây là "chặn trên". Nếu đã vượt quá 1 thì không cho biến hiện tại $True$.

**3. Tại sao Sequential encounter encoding tốt ?**

- **Văn vẻ:**
    - Nếu như ở binomial encoding các biến $X_i$ nói chuyện trực tiếp với nhau $(\neg X_i \land \neg X_j)$ thì ở Sequential encounter encoding lại cho các biến  $X_i$ nói chuyện thông qua các biến trạng thái.
    - Thay vì $X_5$ phải nói chuyện lần lượt với $X_1, X_2, X_3, X_4, ..., X_n$ để bảo `"Nếu tôi nói thì các ông hãy im nhé"` thì $X_5$ chỉ cần hỏi $S_4$ là người đứng ở bên ngoài xem từ $X_1$ đến $X_4$ đã ai `"Nói"` chưa
    - `Hiệu ứng Domino`: Khi $X_2$ được gán là True, nó làm $S_2$ True. $S_2$ lập tức kéo $S_3, S_4, \dots, S_n$ thành $True$ theo một chuỗi phản ứng dây chuyền. Ngay lập tức, tất cả các biến $X$ tương ứng bị "khóa" lại bởi $S_{i-1} \rightarrow \neg X_i (1 < i \le n)$.
- **Độ phức tạp:**
    - Số lượng biến phụ:
        + $n - 1$ biến phụ
        > Tổng $n - 1$ biến phụ hay $O(n)$ biến phụ, trong khi Binomial thì không có biến phụ nào
    - Số lượng clauses (mệnh đề):
        + $X_i \rightarrow S_i (0 < i < n)$ có $n - 1$ clauses
        + $S_{i-1} \rightarrow S_i (1 < i < n)$ có $n-2$ clauses
        + $S_{i-1} \rightarrow \neg X_i (1 < i \le n)$  có $n-1$ clauses
        > Tổng $3n - 4$ clauses hay $O(n)$ clauses, thấp hơn nhiều so với $O(n^2)$ của Binomial encoding

## II. ALO (At least one)

**1. AMO** 
- $X_1 + X_2 + X_3 + ... + X_n \ge 1 $ 


**2. Encoding**


- **Ý tưởng:**
  + Sử dụng biến $S_i$ để lưu trạng thái trước từ $X_i$ chở về trước đã có ít nhất $i$ biến $True$ hay chưa
- **Biến:**
  + n biến chính $X_i (0 < i \le n)$
  + n - 1 biến phụ $S_i (0 < i < n)$
- **Ràng buộc:**
    + `Khởi tạo:`
        * $X_1 \rightarrow S_1$
        * $S_1 \rightarrow X_1$
          > Thiết lập trạng thái ban đầu: Nếu biến đầu tiên đúng, bộ đếm bắt đầu ở mức 1. Cần công thức thứ 2 để tránh trường hợp gán $S_1 = 1$ mà $X_1 = 0$ thì vẫn đúng 
    + `Duy trì trạng thái:`
        * $X_i \rightarrow S_i (1 < i < n)$
        * $S_{i-1} \rightarrow S_i (1 < i < n)$
          > Duy trì trạng thái "đã có ít nhất một" qua các bước từ $i=2$ đến $n-1$.
    + `Ràng buộc để không thêm bit 1 vô căn cứ`
        * $\neg X_i \land \neg S_{i-1} \rightarrow \neg S_i  (1 < i < n)$
          > Đảm bảo $S_i$ không bị gán bằng $1$ vô căn cứ
    + `Ràng buộc loại trừ:` 
        * $S_{n-1} \lor X_n$
          > Cấm trường hợp tất cả đều là $False$.

**3. Độ phức tạp**


 - Số lượng biến phụ:
   + $n - 1$ biến phụ
        > Tổng $n - 1$ biến phụ hay $O(n)$ biến phụ
 - Số lượng clauses (mệnh đề):
    + $X_i \rightarrow S_i (0 < i < n)$ có $n - 1$ clauses
    + $S_1 \rightarrow X_1$ có $1$ clause
    + $S_{i-1} \rightarrow S_i (1 < i < n)$ có $n-2$ clauses
    + $\neg X_i \land \neg S_{i-1} \rightarrow \neg S_i  (1 < i < n)$ có $n-2$ clauses
    + $S_{n-1} \lor X_n$ có $1$ clause
   > Tổng $3n - 3$ clauses hay $O(n)$ clauses

**4. Tại sao ALO lại cần thêm $S_1 \rightarrow X_1$ và $\neg X_i \land \neg S_{i-1} \rightarrow \neg S_i$ mà AMO không cần?**

- Trong AMO việc gán $S_i = 1$ ngay cả khi từ $i$ trở về trước không có bit $1$ nào lại không làm ảnh hưởng đến tính đúng sai của AMO vì:
    + Giả sử nó gán $S_i = 1$ mà chưa thực sự có bit $1$ nào từ $i$ trở về trước thì đến bước $i + 1$ nó sẽ ngừng không thêm bit $1$ nữa vậy thì thực tế chẳng có bit $1$ nào cả nên số bit $1$ vẫn nhỏ hơn $k$ vẫn đúng với AMO
- Nhưng trong ALO việc gán $S_i = 1$ vô căn cứ như trên lại sai vì tại bước $i$ nó phải chắc chắn có ít nhất $1$ bit $1$ thì đến cuối mới có ít nhất $1$ bit 1 thực sự. Nên để tránh solver tự ý gán $S_1 = 1$ mà $X_1$ chưa chắc biết là gì thì phải thêm $S_1 \rightarrow X_1$. Và để đảm bảo nó không tự ý gán $S_i = 1$ một cách vô căn cứ thì phải có $\neg X_i \land \neg S_{i-1} \rightarrow \neg S_i$ (nghĩa là nếu chưa có biến nào $True$ thì trạng thái bây giờ không thể được gán là có ít nhất 1 biến là $True$)

## III. AMK (At most K)

**1. AMK** 
- $X_1 + X_2 + X_3 + ... + X_n \le K $ 


**2. Encoding**

- **Ý tưởng:**
  + $S_{i,j}$ để lưu trạng xem từ $X_i$ trở về trước có `ít nhất` $j$ biến $True$ hay chưa.
  + Để tổng $X_1 + X_2 + ... + X_n \le k \leftrightarrow$ mọi `prefix sum` không vượt quá $k$ **nên** ta sử dụng dạng `>=` để phát hiện sự vượt ngưỡng
- **Biến:**
  + n biến chính $X_i (1 \le i \le n)$
  + $O(nk)$ biến phụ $S_{i,j} (1 \le i < n, 1 \le j \le k)$
    * $S_{i,j} = 1 \leftrightarrow \sum_{t=1}^{i} X_t \geq j $
- **Ràng buộc:**
  + $X_i \rightarrow S_{i,1} (1 \le i \le n -1)$ (1)
  + $S_{i-1,j} \rightarrow S_{i, j} (2 \le i \le n - 1, 1 \le j \le min(i - 1, k))$ (2) 
    > * Tính kế thừa. Nếu ở bước trước đã đếm được $j$ biến đúng, thì ở bước này số lượng biến đúng chắc chắn vẫn ít nhất là $j$
    > * $S_{i-1,j}$ là có tối đa $i-1$ biến $True$, $S_{i,j}$ có tối đa $i$ biến $True$ nhưng để đúng với cả mệnh đề ta lấy $j \le (n-1)$. Hiểu nhanh thì $j$ không bao giờ được vượt quá $k$ hoặc số biến
  + $X_i \land S_{i-1, j-1} \rightarrow S_{i,j} (2 \le i \le n -1, 1 \le j \le min(i, k))$ (3) 
    >* Quy tắc cộng dồn. Nếu ta vừa có thêm một biến $X_i$ đúng và trước đó đã gom được $j-1$ biến đúng, thì bây giờ chắc chắn có $j$ biến đúng
    >*   Có $S_{i-1, j-1}$ nên $j \le i$ và $S_{i,j}$ nên $j \le i$. Vậy tổng kết là lấy $j \le i$
  + $X_i \to \neg S_{i-1,k} (k +1 \le i \le n)$ (8)
    > * Chốt chặn, nếu $X_i$ mà đúng thì từ mốc $i-1$ trở về trước không được đủ $k$ bit 1
    > * Có $S_{i-1, k}$ nên $k \le i - 1 \leftrightarrow i \ge k + 1$

**3. Độ phức tạp**

- Số biến phụ:
  + $O(nk)$ biến phụ
- Số clauses:
  + $O(nk)$ clauses

**4. Tại sao không sử dụng dấu `=` để kiểm tra sự vượt ngưỡng mà lại sự dụng dấu `>=`?**
+ Giả thuyết nếu dùng dấu `=` để kiểm tra sự vượt ngưỡng. Vì ta thấy nếu chạm đến $k$ biến $True$ ở giai đoạn cập nhật nào đó thì sẽ không cho các biến chưa cập nhật là $True$ được nữa.
+ Rất có thể việc sử dụng dấu `=` vẫn được nhưng số mệnh đề sẽ tăng lên. Và với các công thức của sinz không thể hiểu $S_{i,j}$ $True$ khi prefix sum từ 1 đến $i$ bằng $j$ vì các lí do sau đây:
    + Trong một bộ giải SAT các biến bị $S_{i,j}$ đóng vai trò là `bộ nhớ`.
        * Với dấu `>=`: Một khi tổng đã đạt ngưỡng $j$ tại điểm $i$ thì biến $S_{i,j}$ sẽ thành $True$ và giả sử bộ giải thử nghiệm và cho biến $X_{i+1}$ là $True$ thì $S_{i+1,j}$ vẫn là $True$. Rất có ý nghĩa cho việc cập nhật liên tục
        * Với dấu `=`: Nếu ở bước $i$ tổng đã bằng $j$, nhưng sang bước $i + 1$ mà $X_{j+1}$ là $True$ thì với ngữ nghĩa `chính xác bằng` $j$ hay chưa thì $S_{i+1, j}$ lại là $False$. Có nghĩa trạng là không tái sử dụng được ở trạng thái trước mà có tái sử dụng được thì số mệnh đề phải tăng lên
      > Nếu dùng dấu =, bộ nhớ sẽ bị xóa sạch mỗi khi có biến mới là $True$. Không thể dùng một mộ nhớ chập chờn để canh giữ ngưỡng $k$
    +  Cơ chế chốt chặn (The trigger Mechanism). Công thức dàng buộc của AMK là  $\neg X_i \lor \neg S_{i-1, k }$ (nếu đã đạt ngưỡng k, thì biến hiện tại không được phép vượt là $\ True$)
        * Với dấu `>=`: Một khi S đạt ngưỡng k, nó chốt ở trạng thái $True$ mãi mãi. Do đó bất kể bước tiếp theo là bước thứ bao nhiêu, thì nó luôn ngăn cho X là $True$
        * Với dấu `=`: Giả sử solver đang thử nghiệm, tổng vọt lên $k + 1$ lúc này S (nghĩa là = k) sẽ biến thành $False$. Khi S là $False$ chốt $\neg X_i \lor \neg S_{i-1, k }$ sẽ bị vô hiệu hóa vì ($False$ phù định thành $True$ mệnh đề luôn đúng)
       > Hệ thống "mở cửa" cho phép X là $True$ thoải mái vì nó tưởng "À, bây giờ tổng không phải bằng k nữa rồi, chắc là an toàn"

## IV. ALK (At least K)

**1. ALK**
- $X_1 + X_2 + X_3 + ... + X_n \geq K $ 


**2. Encoding**

- **Ý tưởng:**
  + Vẫn sử dụng $S_{i,j}$ với ngữ nghĩa như của AMK
- **Biến:** Tương tự AMK
- **Ràng buộc:**
  + $X_i \rightarrow S_{i,1} (1 \le i \le n -1)$ (1)
  + $S_{i-1,j} \rightarrow S_{i, j} (2 \le i \le n - 1, 1 \le j \le min(i - 1, k))$ (2) 
    > * Tính kế thừa. Nếu ở bước trước đã đếm được $j$ biến đúng, thì ở bước này số lượng biến đúng chắc chắn vẫn ít nhất là $j$
    > * $S_{i-1,j}$ là có tối đa $i-1$ biến $True$, $S_{i,j}$ có tối đa $i$ biến $True$ nhưng để đúng với cả mệnh đề ta lấy $j \le (n-1)$. Hiểu nhanh thì $j$ không bao giờ được vượt quá $k$ hoặc số biến
  + $X_i \land S_{i-1, j-1} \rightarrow S_{i,j} (2 \le i \le n -1, 2 \le j \le min(i, k))$ (3) 
    >* Quy tắc cộng dồn. Nếu ta vừa có thêm một biến $X_i$ đúng và trước đó đã gom được $j-1$ biến đúng, thì bây giờ chắc chắn có $j$ biến đúng
    >*   Có $S_{i-1, j-1}$ nên $j \le i$ và $S_{i,j}$ nên $j \le i$. Vậy tổng kết là lấy $j \le i$
  + $\neg X_i \land \neg S_{i-1, j} \rightarrow \neg S_{i,j} (2 \le i \le n - 1, 0 < j \le min(i - 1, k))$ (4)
    > * Nếu biến hiện tại $X_i = 0$ và bước trước cũng chưa đạt đến $j$ biến đúng, thì bước này tuyệt đối không thể có $j$ biến đúng. Ngăn chặn việc bit $j$ tự bật lên mà không có lí do
    > * Vì có $S_{i-1, j}$ nên $j \le i - 1$ vì có $S_{i,j}$ nên $j \le i$. Vậy $j \le i-1$ 
  + $\neg X_i \land \neg S_{i,i} (1 \le i \le k)$ (5)
    > * Ở bước thứ $i$ nếu $X_i$ sai thì không thể có chuyện toàn bộ $i$ biến đều $True$.
    > * Chỉ xét $i$ đến $k$ vì $max(j) = k$ nên để $j = i$ thì $ i \le k$
  + $\neg S_{i-1, j-1} \rightarrow \neg S_{i,j} (2 \le i \le n -1, 2 \le j \le min(i,k))$ (6)
    > * Nếu chưa gom đủ $j - 1$ ở bước trước thì dù bước hiện tại có $X_i$ đúng đi chăng nữa cũng không đạt được $j$ biến $True$
    > *  Có $S_{i-1, j-1}$ nên $j \le i$ và $S_{i,j}$ nên $j \le i$. Vậy tổng kết là lấy $j \le i$
  + $S_{n-1,k} \lor (S_n \land S_{n-1,k-1}) $ (7) 
    > * TH1: Đã đủ $k$ biến $True$ thì bước trước thì ALK đã xong
    > * TH2: Đạt được $k - 1$ biến $True$ ở bước trước thì biến hiện tại $X_n$ phải $True$
    > * Nếu cả 2 sai thì solver phải chọn các giá trị $X_i$ trước đó sao cho đúng

**3. Tại sao ALK lại sử dụng thêm 3 công thức (4), (5), (6) nữa mà AMK lại không. Và AMK có sử dụng được 3 công thức (4), (5), (6) hay không. Nếu có thì có ảnh hưởng như thế nào đến hiệu năng giải của solver?**

- ALK sử dụng thêm 3 công thức 4,5,6 là vì
    + Với bài toán AMK (Không được quá $k$ bit  $1$):
        * Ràng buộc chính: Công thức (8) nói rằng nếu đến bước $i-1$ đã đủ $k$ bit $1$ rồi thì không được phép thêm bit $1$ ở bước $i$ nữa
        * Nếu không có (4,5,6): Thì $S_{i,j}$ sẽ không bị ép bằng 0 nếu thiếu bit $1$.
        * Nếu Solver gán $S=1$ khi mới chỉ có ít hơn $k$ biến $X$ là $True$ thì sao. `Thì không sai` vì nếu solver gán $S_{i,k} = 1$ mà thực chất chưa đủ $k$ bit 1: Lúc này công thức (8) sẽ chặn không cho các biến sau này là $1$ nữa vậy vấn đúng vì chưa đủ $k$
    + Với bài toán ALK (Tối thiểu $k$ bit $1$):
        * (1) $X_i \to R_{i,1}$: Khi $X_i = 0$, vế trái sai $\implies$ Mệnh đề này luôn đúng, dù $R_{i,1}$ là 0 hay 1.
        * (3) $X_i \wedge R_{i-1,j-1} \to R_{i,j}$: Khi $X_i = 0$, vế trái ($0 \wedge \dots$) chắc chắn sai $\implies$ Mệnh đề này luôn đúng, dù $R_{i,j}$ là 0 hay 1.
        * (2) $R_{i-1,j} \to R_{i,j}$: Đây là mắt xích quan trọng. Nếu trình giải SAT muốn "gian lận" để $R_{n,k} = 1$:
            * Nó chỉ cần gán tất cả các biến phụ $R_{i,j} = 1$.
            * Khi đó, vế trái $R_{i-1,j} = 1$ và vế phải $R_{i,j} = 1$. Theo bảng chân trị: $1 \to 1$ là ĐÚNG.
        *  Kết quả: Trình giải SAT gán $X = 0$ và $R = 1$ cho toàn bộ hệ thống. Kiểm tra lại các công thức (1, 2, 3), tất cả đều thỏa mãn (mang giá trị $True$). Không có "điều gì khác" bị vi phạm ở đây cả!
        *  Vậy nên cần các công thức 4,5,6 để ngăn chặn việc `công nhận ảo`, chưa có ít $k$ bit $1$ mà đã báo hiệu là có $k$ bit $1$.
        
- AMK hoàn toàn sử dụng được 4,5,6 nhưng không cần thiết thì việc gán $S_{i,j} = 1$ trước không ảnh hưởng đến tính đúng sai của bài toán. 
- AMK mà dùng thì bị thừa mệnh đề có ảnh hưởng đến hiệu năng.

## V. Ứng dụng

**1. encoding AMO**

```python
def AMO(g, vars, start_id_var):
    n = len(vars)
    temp_var = list(range(start_id_var, start_id_var + n - 1))
    for i in range (0, n - 1):
        g.add_clause([-vars[i], temp_var[i]])
    for i in range (1, n - 1):
        g.add_clause([-temp_var[i-1], temp_var[i]])
    for i in range (1, n):
        g.add_clause([-temp_var[i-1], -vars[i]])
    return start_id_var + n - 1
```

**2. encoding ALO**

```python
def ALO(g, vars, start_id_var):
    n = len(vars)
    temp_var = list(range(start_id_var, start_id_var + n - 1))
    for i in range (0, n - 1):
        g.add_clause([-vars[i], temp_var[i]])
    for i in range (1, n - 1):
        g.add_clause([-temp_var[i-1], temp_var[i]])
        g.add_clause([vars[i], temp_var[i-1], -temp_var[i]])
        g.add_clause([-temp_var[0], vars[0]])
    g.add_clause([temp_var[n-2], vars[n-1]])
    return start_id_var + n - 1
```

**3. encoding AMK**

```python
def AMK(g, vars, k, start_id_var):
    n = len(vars)
    S = [[start_id_var + i*k + j for j in range(k)] for i in range(n)]

    for i in range(n):
        g.add_clause([-vars[i], S[i][0]])

    for i in range(1, n - 1):
        for j in range(min(i - 1, k) + 1):
            g.add_clause([-S[i-1][j], S[i][j]])

    for i in range(1, n - 1):
        for j in range(1, min(k, i) + 1):
            g.add_clause([-vars[i], -S[i-1][j-1], S[i][j]])

    for i in range(k, n):
        g.add_clause([-vars[i], -S[i-1][k]])

    return start_id_var + n*k
```

**4. encoding ALK**

```python
def ALK(g, vars, k, start_id_var):
    n = len(vars)
    S = [[start_id_var + i*k + j for j in range(k)] for i in range(n)]

    for i in range(n):
        g.add_clause([-vars[i], S[i][0]])

    for i in range(1, n - 1):
        for j in range(min(i - 1, k) + 1):
            g.add_clause([-S[i-1][j], S[i][j]])

    for i in range(1, n - 1):
        for j in range(1, min(k, i) + 1):
            g.add_clause([-vars[i], -S[i-1][j-1], S[i][j]])

    for i in range (1, n - 1):
        for j in range (min(i-1, k) + 1):
            g.add_clause([vars[i], temp_var[i-1][j], -temp_var[i][j]])

    for i in range (k):
        g.add_clause([vars[i], -temp_var[i][i])

    for i in range (1, n - 1):
        for j in range (1, min(i, k) + 1):
            g.add_clause([temp_var[i-1][j-1], temp_var[i][j]])

    return start_id_var + n*k 
```

**5. N_queen**

```python
import math
from pysat.solvers import Glucose3
```

```python
def var_id(r, c, n):
    return r * n + c + 1
```

```python
def solve_n_queen(n):
    
    g = Glucose3()
    next_var = n * n + 1

    # Row constraints
    for r in range(n):
        row_vars = [var_id(r, c, n) for c in range(n)]
        #ALO(g, row_vars)
        next_var = ALO(g, row_vars, next_var)
        next_var = AMO(g, row_vars, next_var)

    # Column constraints
    for c in range(n):
        col_vars = [var_id(r, c, n) for r in range(n)]
        #ALO(g, row_vars) 
        next_var = ALO(g, col_vars, next_var)
        next_var = AMO(g, col_vars, next_var)

    # Main diagonals (r - c)
    for d in range(-n + 1, n):
        diag = []
        for r in range(n):
            c = r - d
            if 0 <= c < n:
                diag.append(var_id(r, c, n))
        if len(diag) > 1:
            next_var = AMO(g, diag, next_var)

    # Anti diagonals (r + c)
    for d in range(2 * n - 1):
        diag = []
        for r in range(n):
            c = d - r
            if 0 <= c < n:
                diag.append(var_id(r, c, n))
        if len(diag) > 1:
            next_var = AMO(g, diag, next_var)

    # Solve
    if g.solve():
        model = set(g.get_model())
        for r in range(n):
            row = ['.'] * n
            for c in range(n):
                if var_id(r, c, n) in model:
                    row[c] = 'Q'
            print(' '.join(row))
    else:
        print("UNSATISFIABLE")
```

```python
solve_n_queen(45)
```

**6. Sudoku**

```python
def var_id_3(r, c, k, n):
    return r*n*n + c*n + k + 1
```

```python
def solve_Sudoku(n, grid):
    g = Glucose3()
    next_var = n * n * n + 1
    s = int(math.sqrt(n))

    for r in range (n):
        for c in range (n):
            if grid[r][c] != 0:
                k = grid[r][c] - 1
                g.add_clause([var_id_3(r,c,k,n)])
    
    # cell constraints
    for r in range (n):
        for c in range (n):
            vars = [var_id_3(r, c, k, n) for k in range (n)]
            next_var = ALO(g, vars, next_var)
            next_var = AMO(g, vars, next_var)
            
    # row constraints
    for k in range (n):
        for r in range (n):
            vars = [var_id_3(r, c, k, n) for c in range (n)]
            next_var = ALO(g, vars, next_var)
            next_var = AMO(g, vars, next_var)
            
    # col constraints
    for k in range (n):
        for c in range (n):
            vars = [var_id_3(r, c, k, n) for r in range (n)]
            next_var = ALO(g, vars, next_var)
            next_var = AMO(g, vars, next_var)
            
    # matrix sqrt(n) * sqrt(n) constraints
    for k in range (n):
        for start_r in range (0, n, s):
            for start_c in range (0, n, s):
                vars = []
                for d in range (n):
                    r = start_r + d//s;
                    c = start_c + d%s;
                    vars.append(var_id_3(r, c, k, n))
                next_var = ALO(g, vars, next_var)
                next_var = AMO(g, vars, next_var)

    if g.solve():
        model = g.get_model()
        for i in range(n):
            for j in range(n):
                for k in range(n):
                    if var_id_3(i, j, k, n) in model:
                        print(k+1, end = ' ')
            print("\n")
    else:
        print("UNSATISFIABLE")
                
            
    
```

```python
grid = [
    [5,3,0,0,7,0,0,0,0],
    [6,0,0,1,9,5,0,0,0],
    [0,9,8,0,0,0,0,6,0],
    [8,0,0,0,6,0,0,0,3],
    [4,0,0,8,0,3,0,0,1],
    [7,0,0,0,2,0,0,0,6],
    [0,6,0,0,0,0,2,8,0],
    [0,0,0,4,1,9,0,0,5],
    [0,0,0,0,8,0,0,7,9],
]
```

```python
solve_Sudoku(9, grid)
```
