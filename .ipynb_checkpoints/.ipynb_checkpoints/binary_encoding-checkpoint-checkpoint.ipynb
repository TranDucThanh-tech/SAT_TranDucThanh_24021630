# Binary encoding

> Thảo luận về cách biểu diễn các bài toán khác bằng SAT (ví dụ:
N-queens).

## I. AMO & ALO

**Ý tưởng hàm AMO trong Binary encoding:**
- Hàm AMO nhận các tham số g, vars, start_var_id:
  + g là Glucose3
  + vars là tập hợp biến đang xét (ở bài N_queen là tất cả biến của 1 hàng, 1 cột, 1 đường chéo)
  + Start_var_id là id đầu tiên của biến phụ (để chánh trùng id với biến chính hoặc các biến phụ đã dùng)

```python
import math
from pysat.solvers import Glucose3
```

```python
def AMO(g, vars, start_var_id):
    n = len(vars)
    if n <= 1:
        return start_var_id
    n_bits = math.ceil(math.log2(n))
    term_vars = [start_var_id + i for i in range (n_bits)]

    for idx, x in enumerate(vars):
        for i in range (n_bits):
            if (idx >> i) & 1: #nếu bit thứ i từ phải sang bằng 1 thì lấy khẳng định của biến phụ
                g.add_clause([-x, term_vars[i]])
            else:
                g.add_clause([-x, -term_vars[i]])
    return start_var_id + n_bits
```

```python
def ALO(g, vars):
    g.add_clause(vars)
```

```python
def var_id(r, c, n):
    return r * n + c + 1
```

## II. N_queens

```python
def solve_n_queen(n):
    
    g = Glucose3()
    next_var = n * n + 1

    # Row constraints
    for r in range(n):
        row_vars = [var_id(r, c, n) for c in range(n)]
        ALO(g, row_vars)
        next_var = AMO(g, row_vars, next_var)

    # Column constraints
    for c in range(n):
        col_vars = [var_id(r, c, n) for r in range(n)]
        ALO(g, col_vars)
        next_var = AMO(g, col_vars, next_var)

    # Main diagonals (r - c)
    for d in range(-n + 1, n):
        diag = []
        for r in range(n):
            c = r - d
            if 0 <= c < n:
                diag.append(var_id(r, c, n))
        if len(diag) > 1:
            next_var = AMO(g, diag, next_var)

    # Anti diagonals (r + c)
    for d in range(2 * n - 1):
        diag = []
        for r in range(n):
            c = d - r
            if 0 <= c < n:
                diag.append(var_id(r, c, n))
        if len(diag) > 1:
            next_var = AMO(g, diag, next_var)

    # Solve
    if g.solve():
        model = set(g.get_model())
        for r in range(n):
            row = ['.'] * n
            for c in range(n):
                if var_id(r, c, n) in model:
                    row[c] = 'Q'
            print(' '.join(row))
    else:
        print("UNSATISFIABLE")
```

```python
solve_n_queen(10)
```

## III. Sudoku

```python
def var_id_3(r, c, k, n):
    return r*n*n + c*n + k + 1
```

```python
def solve_Sudoku(n, grid):
    g = Glucose3()
    next_var = n * n * n + 1
    s = int(math.sqrt(n))

    for r in range (n):
        for c in range (n):
            if grid[r][c] != 0:
                k = grid[r][c] - 1
                g.add_clause([var_id_3(r,c,k,n)])
    
    # cell constraints
    for r in range (n):
        for c in range (n):
            vars = [var_id_3(r, c, k, n) for k in range (n)]
            ALO(g, vars)
            next_var = AMO(g, vars, next_var)
            
    # row constraints
    for k in range (n):
        for r in range (n):
            vars = [var_id_3(r, c, k, n) for c in range (n)]
            ALO(g, vars)
            next_var = AMO(g, vars, next_var)
            
    # col constraints
    for k in range (n):
        for c in range (n):
            vars = [var_id_3(r, c, k, n) for r in range (n)]
            ALO(g, vars)
            next_var = AMO(g, vars, next_var)
            
    # matrix sqrt(n) * sqrt(n) constraints
    for k in range (n):
        for start_r in range (0, n, s):
            for start_c in range (0, n, s):
                vars = []
                for d in range (n):
                    r = start_r + d//s;
                    c = start_c + d%s;
                    vars.append(var_id_3(r, c, k, n))
                ALO(g, vars)
                next_var = AMO(g, vars, next_var)

    if g.solve():
        model = g.get_model()
        for i in range(n):
            for j in range(n):
                for k in range(n):
                    if var_id_3(i, j, k, n) in model:
                        print(k+1, end = ' ')
            print("\n")
    else:
        print("UNSATISFIABLE")
                
            
    
```

```python
grid = [
    [5,3,0,0,7,0,0,0,0],
    [6,0,0,1,9,5,0,0,0],
    [0,9,8,0,0,0,0,6,0],
    [8,0,0,0,6,0,0,0,3],
    [4,0,0,8,0,3,0,0,1],
    [7,0,0,0,2,0,0,0,6],
    [0,6,0,0,0,0,2,8,0],
    [0,0,0,4,1,9,0,0,5],
    [0,0,0,0,8,0,0,7,9],
]
```

```python
solve_Sudoku(9, grid)
```
