# Binary encoding

> Thảo luận về cách biểu diễn các bài toán khác bằng SAT (ví dụ:
N-queens).

## I. N_queens

```python
import math
from pysat.solvers import Glucose3
```

```python
def AMO(g, vars, start_var_id):
    n = len(vars)
    if n <= 1:
        return start_var_id
    n_bits = math.ceil(math.log2(n))
    term_vars = [start_var_id + i for i in range (n_bits)]

    for idx, x in enumerate(vars):
        for i in range (n_bits):
            if (idx >> i) & 1:
                g.add_clause([-x, term_vars[i]])
            else:
                g.add_clause([-x, -term_vars[i]])
    return start_var_id + n_bits
```

```python
def ALO(g, vars):
    g.add_clause(vars)
```

```python
def var_id(r, c, n):
    return r * n + c + 1
```

```python
def solve_n_queen(n):
    from pysat.solvers import Glucose3
    
    g = Glucose3()
    next_var = n * n + 1

    # Row constraints
    for r in range(n):
        row_vars = [var_id(r, c, n) for c in range(n)]
        ALO(g, row_vars)
        next_var = AMO(g, row_vars, next_var)

    # Column constraints
    for c in range(n):
        col_vars = [var_id(r, c, n) for r in range(n)]
        ALO(g, col_vars)
        next_var = AMO(g, col_vars, next_var)

    # Main diagonals (r - c)
    for d in range(-n + 1, n):
        diag = []
        for r in range(n):
            c = r - d
            if 0 <= c < n:
                diag.append(var_id(r, c, n))
        if len(diag) > 1:
            next_var = AMO(g, diag, next_var)

    # Anti diagonals (r + c)
    for d in range(2 * n - 1):
        diag = []
        for r in range(n):
            c = d - r
            if 0 <= c < n:
                diag.append(var_id(r, c, n))
        if len(diag) > 1:
            next_var = AMO(g, diag, next_var)

    # Solve
    if g.solve():
        model = set(g.get_model())
        for r in range(n):
            row = ['.'] * n
            for c in range(n):
                if var_id(r, c, n) in model:
                    row[c] = 'Q'
            print(' '.join(row))
    else:
        print("UNSATISFIABLE")
```

```python
solve_n_queen(10)
```
